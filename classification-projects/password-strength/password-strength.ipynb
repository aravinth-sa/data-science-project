Use NLP techniques

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters

corpus=X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
X=vectorizer.fit_transform(corpus)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

Assigning the scores for each character value

```python
feature_names = vectorizer.get_feature_names_out()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting

# Define OneVsRestClassifier with logistic regression
clf = OneVsRestClassifier(LogisticRegression())

# Fit model
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test,y_test))
```

```python
clf_mb = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf_mb.fit(X_train, y_train)
```

```python
print(clf_mb.score(X_test,y_test))
```

```python
X_predict=np.array(["133333323abcd"])
X_predict=vectorizer.transform(X_predict)
y_pred=clf_mb.predict(X_predict)
print(y_pred)
```

## XGB class

```python
import xgboost as xgb

xgb_clf=xgb.XGBClassifier()
xgb_clf.fit(X_train,y_train)
```

```python
print(xgb_clf.score(X_test,y_test))
```
